# Face Detection Example
In this example we use pre trained weights to
predict faces using opencv built-in dnn module "FaceDetectorYN"


```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
# define parameters for the detector
topk = 5000
nms = 0.3
score = 0.9

# path fo pre-trained nn model weights
detection_model = "face_detection_yunet_2022mar.onnx"

# create detector model
detector = cv2.FaceDetectorYN.create(detection_model,"",(320, 320),score,nms,topk)
```

```python
# read image
img = cv2.imread("data/test_04.jpg")

# define input shape 
detector.setInputSize((img.shape[1], img.shape[0]))

# run prediction
faces= detector.detect(img)

# loop over detected faces and draw rectangle
for face in faces[1]:
    p = (int(face[0]), int(face[1]), int(face[2]), int(face[3]))
    cv2.rectangle(img,p, (0,255,0), 2)
```

```python
# plot the results
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
