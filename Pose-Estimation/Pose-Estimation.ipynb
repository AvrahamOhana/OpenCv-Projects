# Pose Estimation Example
In this example we use pre trained weights to predict body pose using mediapipe

```python
import cv2
import os
import matplotlib.pyplot as plt
import mediapipe as mp
```

```python
# init the pose detector and the draw module
mp_draw = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()
```

```python
# read the test image
img = cv2.imread("test.jpg")

# convert bgr channels to rgb. the detector use rgb format
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# predict the pose
res = pose.process(img_rgb)
```

```python
# if pose found, draw landmarks

if res.pose_landmarks:
    mp_draw.draw_landmarks(img, 
                           res.pose_landmarks,
                           mp_pose.POSE_CONNECTIONS,
                           mp_draw.DrawingSpec(color=(0,0,255)),
                           mp_draw.DrawingSpec(color=(0,255,0))
                           )
    
```

```python
# plot the results
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
